```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.normal(0, 1, 1000)

plt.hist(random_numbers, bins=100, color='gold', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Random Numbers")

plt.savefig("normal_distribution_histogram.pdf")
plt.show()
```
