```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
```

```python
def exp_func(x):
    return np.exp(x)

y = exp_func(x)

plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Plot of $e^x$ from 0 to 1")

# Save the plot as a PDF
plt.savefig("exp_plot.pdf")
plt.show()
```
